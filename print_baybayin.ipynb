```python
def baybay(file):
    """Returns a baybayin character given the filepath"""
    path = ('/mnt/processed/private/msds2022/cpt11/ML3/'
        'Baybayin-Handwritten-Character-Dataset/raw/')
    letter = file.split('/')[0]
    print(f'Baybayin character: {letter}')
    return plt.imshow(cv2.imread(path+file))
```

```python
# gets the files used for test and train

from os import listdir
from os.path import isfile, join

myPath = ('/mnt/processed/private/msds2022/cpt11/ML3/'
'Baybayin-Handwritten-Character-Dataset/raw/')

# All the file paths will be stored in this list
filesList= []

for path, subdirs, files in os.walk(myPath):
    for name in files:
        filesList.append(os.path.join(name))
        
filesList
```

```python
baybay('a/a_00001_file001.jpg');
```

```python
#print a random baybayin character

def baybay_random():
    """Returns a random baybayin character"""
    import random
    rand_path = random.choice(filesList)
    rand_dir = rand_path.split('_')[0]
    random_path = rand_dir + '/' + rand_path
    print(random_path)
    
    return baybay(random_path);
```
